```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

job_data = pd.read_csv('/kaggle/input/linkedin-job-postings/postings.csv')
job_data = job_data[['job_id', 'company_name', 'title', 'description']].dropna()

# Generate embeddings
job_descriptions = job_data['description'].tolist()
job_embeddings = model.encode(job_descriptions, show_progress_bar=True)

np.save('job_embeddings.npy', job_embeddings)
job_data.to_csv('job_data.csv', index=False)
print("Saved job embeddings and metadata.")
```
