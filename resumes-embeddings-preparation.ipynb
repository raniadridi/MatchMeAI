```python
import os
import PyPDF2
import numpy as np
from sentence_transformers import SentenceTransformer

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = []
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text.append(page_text)
        return " ".join(text)

def process_all_pdfs_in_folder(root_folder):
    file_texts = []  
    file_paths = []  

    for subdir, dirs, files in os.walk(root_folder):
        for file in files:
            if file.endswith(".pdf"):
                file_path = os.path.join(subdir, file)
                try:
                    text = extract_text_from_pdf(file_path)
                    if text.strip():
                        file_texts.append(text)
                        file_paths.append(file_path)
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

    resume_embeddings = model.encode(file_texts)
    return file_texts, file_paths, resume_embeddings

root_folder = '/kaggle/input/resume-dataset/data/data'
resume_texts, resume_files, resume_embeddings = process_all_pdfs_in_folder(root_folder)

np.save('resume_embeddings.npy', resume_embeddings)
np.save('resume_texts.npy', np.array(resume_texts))
np.save('resume_files.npy', np.array(resume_files))
print("Saved resume embeddings and metadata.")
```
